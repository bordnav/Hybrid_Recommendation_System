## Collaborative Filtering based Recommender System

```python
ratings_with_name
```

```python
ratings_with_name.groupby('User-ID').count()
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
users_who_review = x[x].index
```

```python
ratings_with_name['User-ID'].isin (users_who_review)
```

```python
filtered_rating = ratings_with_name [ratings_with_name['User-ID'].isin (users_who_review)]
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
y[y]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
famous_books
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt
```

```python
pt.fillna(0,inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

```python
similarity_score.shape
```

```python
list(enumerate(similarity_score[0]))
```

```python
sorted(list(enumerate(similarity_score[3])), key =lambda x:x[1], reverse = True)
```

```python
sorted(list(enumerate(similarity_score[3])), key =lambda x:x[1], reverse = True)[1:6]
```

```python
def recommend(book_name) :
    #index fetch from the array of similarity scores
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key =lambda x:x[1], reverse = True)[1:6]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
np.where(pt.index=='Zoya')[0][0]
```

```python
recommend('Nights in Rodanthe')
```
